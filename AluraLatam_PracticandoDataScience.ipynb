### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
df = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

df['Fecha de Compra'] = pd.to_datetime(df['Fecha de Compra'], dayfirst=True)

df['Facturacion'] = df['Precio'] + df['Costo de envío']

print("Facturación total: ", df['Facturacion'].sum())
print("Facturación promedio por compra: ", df['Facturacion'].mean())

facturacion_por_categoria = df.groupby('Categoría del Producto')['Facturacion'].sum().sort_values(ascending=False)
print("\nFacturación por categoría de producto:")
print(facturacion_por_categoria)

df['Mes'] = df['Fecha de Compra'].dt.to_period('M')
facturacion_por_mes = df.groupby('Mes')['Facturacion'].sum()
print("\nFacturación por mes:")
print(facturacion_por_mes)

top_productos = df.groupby('Producto')['Facturacion'].sum().sort_values(ascending=False).head(10)
print("\nTop 10 productos por facturación:")
print(top_productos)
```

# 2. Ventas por categoría

```python
ventas_por_categoria = df.groupby('Categoría del Producto').size().sort_values(ascending=False)
print("Cantidad de ventas por categoría:\n", ventas_por_categoria)

facturacion_por_categoria = df.groupby('Categoría del Producto')['Facturacion'].sum().sort_values(ascending=False)
print("\nFacturación por categoría:\n", facturacion_por_categoria)
```

# 3. Calificación promedio de la tienda


```python
tienda["Tienda"] = "Tienda 1"
tienda2["Tienda"] = "Tienda 2"
tienda3["Tienda"] = "Tienda 3"
tienda4["Tienda"] = "Tienda 4"

df = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

calificacion_promedio_tienda = df.groupby("Tienda")["Calificación"].mean()
print("Calificación promedio por tienda:\n", calificacion_promedio_tienda)
```

# 4. Productos más y menos vendidos

```python
productos_vendidos = df['Producto'].value_counts()

print("Top 10 productos más vendidos:\n", productos_vendidos.head(10))
print("\nTop 10 productos menos vendidos:\n", productos_vendidos.tail(10))
```

# 5. Envío promedio por tienda

```python
envio_promedio_tienda = df.groupby("Tienda")["Costo de envío"].mean()
print("Envío promedio por tienda:\n", envio_promedio_tienda)
```
